```python
columns_to_keep = ['game_id','home_club_id', 'away_club_id', 'home_club_goals', 'away_club_goals',
                   'home_club_position', 'away_club_position', 'home_club_name',
                   'away_club_name', 'aggregate']


df_filtered = games[columns_to_keep]


print(df_filtered.head())
```

```python
import pandas as pd


columns_to_keep = ['game_id', 'hosting', 'is_win']


df_filtered2 = club_games[columns_to_keep]


print(df_filtered2.head())
```

```python
print("Duplicates in df_filtered:")
print(df_filtered[df_filtered.duplicated('game_id', keep=False)])

print("Duplicates in df_filtered2:")
print(df_filtered2[df_filtered2.duplicated('game_id', keep=False)])


df_filtered_unique = df_filtered.drop_duplicates(subset='game_id', keep='first')
df_filtered2_unique = df_filtered2.drop_duplicates(subset='game_id', keep='first')


merged_df = pd.merge(df_filtered_unique, df_filtered2_unique, on='game_id', how='inner')


print("Merged DataFrame:")
print(merged_df.head())


print("Shape of the merged DataFrame:")
print(merged_df.shape)
```

```python
merged_df['points_home'] = 0
merged_df['points_away'] = 0


merged_df.loc[(merged_df['home_club_goals'] > merged_df['away_club_goals']), 'points_home'] = 3
merged_df.loc[(merged_df['home_club_goals'] < merged_df['away_club_goals']), 'points_away'] = 3


merged_df.loc[(merged_df['home_club_goals'] == merged_df['away_club_goals']), 'points_home'] = 1
merged_df.loc[(merged_df['home_club_goals'] == merged_df['away_club_goals']), 'points_away'] = 1

print(merged_df.head())
```

```python
home_points = merged_df.groupby('home_club_id')['points_home'].sum().reset_index()
home_points.rename(columns={'home_club_id': 'club_id', 'points_home': 'total_points'}, inplace=True)


away_points = merged_df.groupby('away_club_id')['points_away'].sum().reset_index()
away_points.rename(columns={'away_club_id': 'club_id', 'points_away': 'total_points'}, inplace=True)


total_points = pd.concat([home_points, away_points], axis=0)
total_points = total_points.groupby('club_id')['total_points'].sum().reset_index()


total_points.rename(columns={'club_id': 'team_id', 'total_points': 'total_points'}, inplace=True)


merged_df = pd.merge(merged_df, total_points, left_on='home_club_id', right_on='team_id', how='left')
merged_df.rename(columns={'total_points': 'home_team_total_points'}, inplace=True)


merged_df = pd.merge(merged_df, total_points, left_on='away_club_id', right_on='team_id', how='left')
merged_df.rename(columns={'total_points': 'away_team_total_points'}, inplace=True)


merged_df.drop(columns=['team_id_x', 'team_id_y'], inplace=True)


print(merged_df.head())
```

```python
merged_df.isnull().sum()
```

```python
import pandas as pd

df = merged_df
df.dropna(inplace=True)


print(df.head())
print("Shape of the cleaned DataFrame:", merged_df.shape)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
plt.hist(df['home_club_goals'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Home Club Goals')
plt.xlabel('Home Club Goals')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(merged_df['away_club_goals'], bins=20, color='salmon', edgecolor='black')
plt.title('Distribution of Away Club Goals')
plt.xlabel('Away Club Goals')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay


y_true = merged_df['is_win']
y_pred = (merged_df['points_home'] > merged_df['points_away']).astype(int)


conf_matrix = confusion_matrix(y_true, y_pred)


disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=['Not Win', 'Win'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix of Match Outcomes')
plt.show()
```

```python
import pandas as pd


df['total_points'] = df['home_team_total_points'] + df['away_team_total_points']


df['rank'] = df['total_points'].rank(ascending=False, method='dense')


df['top_16'] = df['rank'].apply(lambda x: 1 if x <= 16 else 0)


print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")


top_16_teams = df[df['top_16'] == 1]


top_16_teams = top_16_teams.sort_values(by='total_points', ascending=False)


plt.figure(figsize=(10, 6))
sns.barplot(x='total_points', y='home_club_name', data=top_16_teams, palette='Blues_d')


plt.title('Top 16 Teams by Total Points', fontsize=16)
plt.xlabel('Total Points', fontsize=12)
plt.ylabel('Club Name', fontsize=12)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

numerical_df = df.select_dtypes(include=[np.number])


plt.figure(figsize=(15, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Dataset Features")
plt.show()
```

```python
features = df[['home_club_position', 'away_club_position','home_team_total_points', 'away_team_total_points','total_points']]

target = df['top_16']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)


y_pred_dt = dt.predict(X_test)


accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f'Decision Tree Accuracy: {accuracy_dt}')


conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)
print("Confusion Matrix:")
print(conf_matrix_dt)


class_report_dt = classification_report(y_test, y_pred_dt)
print("Classification Report:")
print(class_report_dt)
```

```python
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)
print("Confusion Matrix:")
print(conf_matrix_dt)


plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_dt, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Not Top 16', 'Top 16'],
            yticklabels=['Not Top 16', 'Top 16'])
plt.title('Confusion Matrix', fontsize=16)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.show()
```
